```python
#Importamos o algoritimo
from sklearn.cluster import KMeans
#Isso aqui é bem legal, para fazer transformação nos dados
from sklearn.preprocessing import LabelEncoder
#Importamos um conjunto de dados para trabalhar
#O seaborn traz os dados em um formato bem melhor
import seaborn as sns
#Fazer os plots
import matplotlib.pyplot as plt

```

Se você já trabalha ou pesquisa sobre ciência de dados em algum momento já deve ter esbarrado com o dataset Ires.
O nosso problema aqui é bem simples, as caracteristicas de uma flor Ires, queremos classificar ela em um dos 3 grupos: Versicolor, Setosa e Virginica. Não vamos entrar em detalhes sobre ele, mas pode saber mais explorando o dataset e lendo sobre ele aqui https://archive.ics.uci.edu/ml/datasets/Iris

```python
# preparando nossos dados
iris = sns.load_dataset('iris')
df = iris.drop('species', axis=1)

# inicializando o kmeans com a quantidade que queremos de clusters
kmeans = KMeans(n_clusters=3)

# Executando o K-means
iris['clusters'] = kmeans.fit_predict(df)

# transformando as especies de iris em numeros para colorir o gráfico
iris['species_encoded'] = LabelEncoder().fit_transform(iris['species'])






```

```python
plt.figure(figsize=(15,5))

plt.subplot(1, 2, 1)
plt.title('Antes')
plt.xlabel('comprimento da pétala')
plt.ylabel('largura da pétala')
plt.scatter(iris['petal_length'], iris['petal_width'], c=iris['species_encoded'])


plt.subplot(1, 2, 2)
plt.title('Depois')
plt.xlabel('comprimento da pétala')
plt.scatter(iris['petal_length'], iris['petal_width'], c=iris['clusters'])

plt.show()

```
